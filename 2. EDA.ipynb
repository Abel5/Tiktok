#### **video_duration_sec**

Create a box plot to examine the spread of values in the `video_duration_sec` column.

````python
# Create a boxplot to visualize distribution of `video_duration_sec`
# Set up the matplotlib figure
plt.figure(figsize=(12, 5))
# Create a box plot to examine the spread of values in the "video_duration_sec" column
sns.boxplot(x=tiktok_data['video_duration_sec'])
# Set the title of the plot
plt.title('Box plot of Video Duration (Seconds)')
````

**Observations from the Box plot**:
**Median** (Line inside the box): The median value is around 32 seconds, which means that 50% of the videos have a duration less than 32 seconds and 50% have a duration more than 32 seconds.
**Interquartile Range** (IQR - The box): The box represents the interquartile range (IQR), which is the range between the first quartile (25th percentile) and the third quartile (75th percentile). It appears to range from approximately 18 seconds to 47 seconds. This range contains the middle 50% of the values of the video_duration_sec column.
**Whiskers** (Lines extending from the box): The whiskers extend from the edges of the box to the minimum and maximum values within 1.5 times the IQR from the first and third quartiles, respectively.
**Outliers** (Points beyond the whiskers): There are no points beyond the whiskers in this plot, suggesting there are no outliers in the video_duration_sec column based on the 1.5*IQR criterion.

**Conclusion**:
The video_duration_sec column has a relatively symmetrical distribution with no observed outliers, and the majority of the videos have a duration between 18 and 47 seconds.

Create a histogram of the values in the `video_duration_sec` column to further explore the distribution of this variable.

````python
# Create a histogram
# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Create a histogram of the values in the "video_duration_sec" column
sns.histplot(tiktok_data['video_duration_sec'], bins=30, kde=True)

# Set the title and labels of the plot
plt.title('Histogram of Video Duration (Seconds)')
plt.xlabel('Video Duration (Seconds)')
plt.ylabel('Frequency')
````

The histogram provides a view of the distribution of the video_duration_sec column, which represents the duration of the videos in seconds. The Kernel Density Estimate (KDE) line gives a smooth curve representing the probability density of the variable.

**Observations from the Histogram**:

**Multimodal Distribution**: The histogram reveals a multimodal distribution with five apparent peaks, indicating that there are several different groups or categories within the video durations.
**Peaks Locations**:
   * **First Peak**: Around 10 seconds, indicating a high frequency of videos with durations in this range.
   * **Second Peak**: Around 20 seconds, revealing another group of videos with durations centered around this value.
   * **Third Peak**: Around 30 seconds, showing another prevalent video duration on the platform.
   * **Fourth Peak**: Around 40 seconds, indicating another common duration for videos.
   * **Fifth Peak**: Between 50 and 60 seconds, representing the last identified peak of common video durations.
   * **Valleys between Peaks**: The valleys (low points) between the peaks represent durations that are less common. These points are approximately at 15 seconds, 25 seconds, 35 seconds, and 45 seconds.
   
**Frequency**: The y-axis represents the frequency of videos. The frequency of videos is notably higher at the identified peak durations compared to the valleys between them.

**Kernel Density Estimate (KDE)**: The smooth KDE line visually reinforces the locations of the peaks and valleys in the distribution, serving as a guide to understanding the probability density of the variable.

**Conclusion**:
The distribution of the video_duration_sec column is multimodal, revealing several preferred video durations on the platform, specifically around 10, 20, 30, 40, and 50-60 seconds. This multimodal pattern might reflect user preferences, content creation trends, or the influence of the platform's features and constraints on video length. The presence of multiple peaks suggests that content creators may be tailoring their videos' lengths to fit these common durations, which could be related to user engagement, platform algorithms, or other factors impacting video visibility and reach.

Create a histogram of the values in the `video_view_count` column to further explore the distribution of this variable.

````python
# Create a histogram
# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Create a histogram of the values in the "video_view_count" column
# Using logarithmic scale for better visualization of the distribution
sns.histplot(np.log1p(tiktok_data['video_view_count']), bins=60, kde=True)

# Set the title and labels of the plot
plt.title('Histogram of Log Transformed Video View Count')
plt.xlabel('Log Transformed Video View Count')
plt.ylabel('Frequency')
````

The histogram represents the distribution of the **video_view_count** column, and to better visualize the spread and skewness of the data, a logarithmic transformation (log1p, log of one plus the value) is applied to the view counts.

**Observations from the Histogram**:
**Right-Skewed Distribution**: The histogram of the log-transformed video_view_count reveals a right-skewed distribution, indicating that the majority of the videos have lower view counts, with fewer videos achieving higher view counts.
**Peak**: The peak of the distribution is visible on the left side of the histogram, indicating the mode of the log-transformed view counts. This peak corresponds to the most common view count range in the original scale.

Tail: The long tail to the right represents videos with higher view counts in the original scale, highlighting the presence of videos with exceptionally high views compared to the majority.

Kernel Density Estimate (KDE): The smooth KDE line provides a visual guide to understanding the probability density of the variable, reinforcing the observations of the skewness and the mode of the distribution.

Conclusion:
The video_view_count column has a right-skewed distribution, indicating a concentration of videos with lower view counts and a decline in the frequency of videos with higher view counts. The long tail to the right in the log-transformed histogram signifies the presence of videos with exceptionally high view counts, which are rare compared to videos in the common view count range. This distribution is typical in social media platforms where a few videos receive a large number of views, while the majority receive fewer views.

### **Task 4. Determine outliers**

When building predictive models, the presence of outliers can be problematic. For example, if you were trying to predict the view count of a particular video, videos with extremely high view counts might introduce bias to a model. Also, some outliers might indicate problems with how data was captured or recorded.

The ultimate objective of the TikTok project is to build a model that predicts whether a video is a claim or opinion. The analysis you've performed indicates that a video's engagement level is strongly correlated with its claim status. There's no reason to believe that any of the values in the TikTok data are erroneously captured, and they align with expectation of how social media works: a very small proportion of videos get super high engagement levels. That's the nature of viral content.

Nonetheless, it's good practice to get a sense of just how many of your data points could be considered outliers. The definition of an outlier can change based on the details of your project, and it helps to have domain expertise to decide a threshold. You've learned that a common way to determine outliers in a normal distribution is to calculate the interquartile range (IQR) and set a threshold that is 1.5 * IQR above the 3rd quartile.

In this TikTok dataset, the values for the count variables are not normally distributed. They are heavily skewed to the right. One way of modifying the outlier threshold is by calculating the **median** value for each variable and then adding 1.5 * IQR. This results in a threshold that is, in this case, much lower than it would be if you used the 3rd quartile.

Write a for loop that iterates over the column names of each count variable. For each iteration:
1. Calculate the IQR of the column
2. Calculate the median of the column
3. Calculate the outlier threshold (median + 1.5 * IQR)
4. Calculate the numer of videos with a count in that column that exceeds the outlier threshold
5. Print "Number of outliers, {column name}: {outlier count}"

```
Example:
Number of outliers, video_view_count: ___
Number of outliers, video_like_count: ___
Number of outliers, video_share_count: ___
Number of outliers, video_download_count: ___
Number of outliers, video_comment_count: ___
```

````python
# Define the list of count variable column names
count_columns = ['video_view_count', 'video_like_count', 'video_share_count', 'video_download_count', 'video_comment_count']

# Initialize a dictionary to store the results
outlier_counts = {}

# Iterate over the column names of each count variable
for column in count_columns:
    # Calculate the IQR of the column
    Q1 = tiktok_data[column].quantile(0.25)
    Q3 = tiktok_data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculate the median of the column
    median = tiktok_data[column].median()
    
    # Calculate the outlier threshold (median + 1.5 * IQR)
    outlier_threshold = median + 1.5 * IQR
    
    # Calculate the number of videos with a count in that column that exceeds the outlier threshold
    outlier_count = (tiktok_data[column] > outlier_threshold).sum()
    
    # Store the result in the dictionary
    outlier_counts[column] = outlier_count

# Print the results
outlier_counts_str = "\n".join([f"Number of outliers, {column}: {count}" for column, count in outlier_counts.items()])
print(outlier_counts_str)
````

**Explanation**:
For each specified column, the Interquartile Range (IQR) and the median are calculated.
The outlier threshold is then calculated as **median+1.5×IQR**

Finally, the number of videos with a count in that column exceeding the outlier threshold is calculated and printed.
These outlier counts give an indication of the number of videos that have exceptionally high counts in each of the respective categories, which can be useful for understanding the distribution of these variables and for identifying extreme cases in the dataset.
